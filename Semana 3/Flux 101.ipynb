```julia
using Flux
```

# Las bases de flux

```julia
actual(x)=4x+2
```

```julia
 x_train, x_test = hcat(0:5...), hcat(6:10...)
y_train, y_test = actual.(x_train), actual.(x_test)
```

```julia
model = Dense(1 => 1)
```

```julia
model.weight
```

```julia
model.bias
```

```julia
loss(x,y)=Flux.Losses.mse(model(x),y)
```

```julia
loss(x_train, y_train)
```

```julia
opt=Descent()
```

```julia
data = [(x_train, y_train)]
```

```julia
parameters = Flux.params(model)
```

```julia
using Flux: train!
```

```julia
train!(loss, parameters, data, opt)
```

```julia
loss(x_train, y_train)
```

```julia
lossT=[223.4126f0,212.03987f0]
```

```julia
append!(lossT,201.2467f0)
```

```julia
for epoch in 1:200
     train!(loss, parameters, data, opt)
     append!(lossT,loss(x_train, y_train))
end
```
